```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import openpyxl
from openpyxl import load_workbook
```

```python
#leemos los datos qe se encuentran en el excel y asignamos la hoja "C1 Milagros"
datos = pd.read_excel("02 CH Genes ordenados con vecinos.xlsx", sheet_name='C1 Milagros')
valores = datos["Unnamed: 4"].values #asignamos los valores de los genes
lista1 = valores.tolist() #hacemos una lista
lista1.pop(0) #eliminamos la primera variable
del lista1[-11:] #eliminamos las 11 ultimas ya que no son genes

#hacemos lo mismo que el primero pero en este caso usaremos la hoja "C4 Jerica"
datos2 = pd.read_excel("02 CH Genes ordenados con vecinos.xlsx", sheet_name='C4 Jerica')
valores = datos2["Unnamed: 4"].values
lista2 = valores.tolist()
lista2.pop(0)
del lista2[-16:] #en este caso teniamos que eliminar las 16 ultimas
```

```python
df = pd.DataFrame({'GEN ABREV': lista1}) #creamos un data frame con los genes de la lista1
archivo11 = load_workbook("TablaGen.xlsx") #insertamos los valores en un excel vacio
ws=archivo11['Hoja1'] #asignamos la hoja que vamos a utilisar

#guardamos los datos del archivo11 en el excel que vamos a llenar
for index, row in df.iterrows(): 
    cell = 'B%d'  % (index + 3)
    ws[cell] = row[0]
archivo11.save("TablaGen.xlsx")

#hacemos lo mismo que el anterior pero esta vez usaremos los datos de la lista2 
#agregandolos en una hoja diferente, que en este caso seria la hoja2
df2 = pd.DataFrame({'GEN ABREV': lista2})
archivo22 = load_workbook("TablaGen.xlsx")
ws2=archivo22['Hoja2']
for index, row in df2.iterrows():
    cell = 'B%d'  % (index + 3)
    ws2[cell] = row[0]
archivo22.save("TablaGen.xlsx")
```

```python
m=3
for i in lista1:
    print(m-2,":",i)
    URL=f"https://www.uniprot.org/uniprot/?query={i}+capra+hircus&sort=score"  #copiamos el URL donde buscaremos el Gen
    page=requests.get(URL)
    soup=BeautifulSoup(page.content,"html.parser")

    link_entry=soup.find_all("td",class_="entryID")
    indice=[]
    for P in link_entry:
        indice.append(P.text)
    try:
        Fin=indice[0]
        url=f"https://www.uniprot.org/uniprot/{Fin}"
        page=requests.get(url)
        soup=BeautifulSoup(page.content,"html.parser")
        
        #Comenzamos a buscar la informacion para llenar las columnas que necesitamos 
        
        #Entry
        dosprimerascol =soup.find_all("h2",class_="page-title")
        for Q in dosprimerascol:
            entry1=Q.text
        entry1=entry1.split()
        Entry=entry1[2]
        
        #Entry_name
        Entry_name="".join(list(entry1[3])[1:len(list(entry1[3]))-1])

        #Protein
        protein=soup.find_all("h1",property="name")
        for Q in protein:
            Protein=Q.text

        #Gen
        gen_abrev=soup.find_all("div",class_='entry-overview-content',id="content-gene")
        for Q in gen_abrev:
            Gen_abrev=Q.text

        #Organismo
        organismo=soup.find_all("div",class_='entry-overview-content',id="content-organism")
        for Q in organismo:
            Organismo=Q.text

        #Status
        status=soup.find_all("span",class_='context-help bin-score tooltipped-click')
        for Q in status:
            STATUS_PRE=Q.text
        Status=STATUS_PRE.split("\n")[0].split(":")[1]

        #Bio_process
        bio_process=soup.find_all("a",onclick="window.ga('UniProt-Entry-View', 'click', 'Display-GO-Term');")
        BIO_PROCESS=[]
        for Q in bio_process:
            BIO_PROCESS.append(Q.text)
        Bio_process=str(BIO_PROCESS)

    #asignamos un except para que cuando encuentre un dato vacio, coloque "na"
    except:
        Entry="na"
        Entry_name="na"
        Protein="na"
        Gen_abrev="na"
        Organismo="na"
        Status="na"
        Bio_process="na"
    
    excel=openpyxl.load_workbook('TablaGen.xlsx') #abrimos el excel donde colocaremos los datos
    sheet=excel['Hoja1'] #seleccionamos la hoja donde los introduciran
    
    #colocamos y asignamos a que columna va a ir cada entidad
    sheet.cell(row=m,column=3).value = Entry
    sheet.cell(row=m,column=4).value = Entry_name
    sheet.cell(row=m,column=5).value = Protein
    sheet.cell(row=m,column=6).value = Gen_abrev
    sheet.cell(row=m,column=7).value = Organismo
    sheet.cell(row=m,column=8).value = Status
    sheet.cell(row=m,column=9).value = Bio_process
    excel.save('TablaGen.xlsx')
    m=m+1
```
